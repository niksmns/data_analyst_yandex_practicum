## Для каждой книги посчитайте количество обзоров и среднюю оценку;

```python
query = (
'''
SELECT b.book_id,
       b.title,
       COUNT(DISTINCT rv.review_id) AS rewiew_count,
       AVG(rt.rating) AS avg_ratings
FROM books AS b
LEFT JOIN reviews AS rv ON rv.book_id = b.book_id
LEFT JOIN ratings AS rt ON rt.book_id = b.book_id
GROUP BY b.book_id
ORDER BY 3 DESC, 4 DESC;
'''
)
pd.io.sql.read_sql(query, con = engine) 
```

## Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры

```python
query = (
'''
SELECT p.publisher,
       COUNT(b.book_id) AS book_total
FROM books AS b
LEFT JOIN publishers AS p ON p.publisher_id = b.publisher_id
WHERE b.num_pages > 50
GROUP BY p.publisher
ORDER BY book_total DESC
LIMIT(1);
'''
)
pd.io.sql.read_sql(query, con = engine) 
```

## Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками

```python
query = (
'''
WITH

books_review AS(
    SELECT b.book_id
    FROM books AS b
    LEFT JOIN ratings AS rt ON b.book_id = rt.book_id
    GROUP BY b.book_id
    HAVING COUNT(DISTINCT rt.rating_id) >= 50
),

ratings_author AS(
    SELECT a.author,
            ROUND(AVG(rt.rating), 2) AS avg_ratings
     FROM books AS b
     JOIN books_review br ON br.book_id = b.book_id
     LEFT JOIN authors AS a ON b.author_id = a.author_id
     LEFT JOIN ratings AS rt ON b.book_id = rt.book_id
     GROUP BY a.author
)

SELECT *
FROM ratings_author
ORDER BY 2 DESC
LIMIT(1);

'''
)
pd.io.sql.read_sql(query, con = engine) 
```

## Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок

```python
query = (
'''
WITH

users AS(
    SELECT rv.username,
           COUNT(DISTINCT rt.rating_id) AS rating_count
    FROM reviews AS rv
    LEFT JOIN ratings AS rt ON rv.username = rt.username
    GROUP BY rv.username
    HAVING COUNT(DISTINCT rt.rating_id) > 50
),

reviews AS(
    SELECT rv.username,
            COUNT(DISTINCT rv.review_id) AS reviews_count
     FROM reviews rv
     JOIN users AS us ON us.username = rv.username
     GROUP BY rv.username
)

SELECT ROUND(AVG(reviews_count)) AS avg_review
FROM reviews

'''
)
pd.io.sql.read_sql(query, con = engine) 
```
